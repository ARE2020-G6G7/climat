```python
print(" On veut faire tourner l'Equation de KAYA - afin de réduire le taux de CO²")
print(" CO² = p*P x pm*PIB/P x ie*TEP/PIB x ip*CO²/TEP")

print(" Consignes de saisie des 5 données /BAISSE : -valeur /HAUSSE : valeur /1 seul INCONNU = 0 ")
c = int(input("On souhaite faire baisser la production de CO2: c à %? --> "))
p = int(input("1- quelle est la variation de la population P en %? --> "))
ip = int(input("2- quelle est la variation de CO2/TEP: ip en %? -->  "))
ie = int(input("3- quelle est la variation de TEP/PIB: ie en %? --> "))
pm = int(input("4- quelle est la variation de PIB/P: pm en %? --> "))


if ip == 0:
    r = c/(p*pm*ie)
    res = "ip"
elif ie == 0:
    r = c/(p*pm*ip)
    res = "ie"
elif pm == 0:
    r = c/(p*ip*ie)
    res = "pm"
elif p == 0:
    r = c/(ie*ip*pm)
    res = "p"
else: 
    r = p*pm*ie*ip
    res = "c"
    
print("Voici les valeurs données de p,ip,ie,pm,c sont",p,",",ip,",",ie,",",pm,",",c,",")
print ("la valeur de r est ",r)
if abs(r)<1:
   print ("La valeur de l'inconnu ",res," est en baisse de ",(1-abs(r))*100,"%")
else: 
   print ("La valeur de l'inconnu ",res," est en hausse de ",(abs(r)-1)*100,"%")
```
